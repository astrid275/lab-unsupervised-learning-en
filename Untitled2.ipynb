# Remove Outliers

```python
def drop_outliers_iqr(df, cols, k=1.5): # K is the constant tuquays method
    X = df[cols]
    Q1 = X.quantile(0.25)
    Q3 = X.quantile(0.75)
    IQR = Q3 - Q1
    mask = ~((X < (Q1 - k*IQR)) | (X > (Q3 + k*IQR))).any(axis=1) 
    return df[mask]


#Q1 = X.quantile(0.25)
#Q3 = X.quantile(0.75)
#IQR = Q3 - Q1

#lower_bound = Q1 - 1.5 * IQR
#upper_bound = Q3 + 1.5 * IQR

#X_no_outliers = X[~((X < (Q1 - 1.5 * IQR)) | (X > (Q3 + 1.5 * IQR))).any(axis=1)]
```
